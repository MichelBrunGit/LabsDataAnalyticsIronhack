```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from math import pi
```

```python
diamonds = pd.read_csv(r'C:\Users\Michel\git2\LabsDataAnalyticsIronhack\module_2\Project_Week_4\diamonds_clean.csv')
diamonds.head()
```

```python
diamonds.sort_values(by=["x", "y"], inplace=True)

line_mpl = plt.figure() 

plt.plot("x","y", data=diamonds, color='gold')

plt.title('Length vs. Width of Diamonds')
plt.xlabel('Lenght, x [mm]')
plt.ylabel('Width, y [mm]')
plt.grid()
plt.legend()
plt.show()
```

```python
quantcut = diamonds.groupby(['cut'])['carat'].agg(['count'])
quantcut.reset_index(inplace=True)
quantcut
```

```python
bar_mpl = plt.figure()
plt.barh("cut","count", data=quantcut,edgecolor='red')

plt.title('Quantity vs Cut')
plt.xlabel('Cut')
plt.ylabel('Quantity')
plt.grid()
plt.show()
```

```python
plt.figure(figsize = [14,10])
plt.scatter(data = diamonds , x = 'carat', y = 'price', alpha = 0.5)

plt.title("Carats vs Price\n", fontweight='bold')
plt.xlabel("Carats")
plt.ylabel("Price")

plt.show()
```

```python
box_mpl = plt.figure()

data=diamonds[['x','y']]
labels = ['x - length', 'y - width']


bplot1 = plt.boxplot(data,
                     vert=True,  
                     patch_artist=True,  
                     labels=labels,
                     ) 

plt.grid(axis='y')
plt.ylabel('Dimmension [mm]')
plt.xlabel('Measurement')

plt.show()
```

```python
CutPrice = diamonds.groupby('cut').price.mean()
CutPrice.plot(kind='pie',figsize=(7,7))
```

```python
sns.set_theme(style="darkgrid")
```

```python
diamonds.sort_values(by=["x", "z"], inplace=True)

sns.lineplot(x="x", y="y",data=diamonds)

plt.title('Length vs. Width')
plt.xlabel('Lenght, x [mm]')
plt.ylabel('Width, z [mm]')
plt.grid()
plt.show()
```

```python
quantity_carat = diamonds.groupby(['carat'])['carat'].agg(['count'])
quantity_carat.reset_index(inplace=True)
quantity_carat
```

```python
ax = sns.barplot(x="carat", y="count", data=quantity_carat)
plt.title('Quantity vs carats')
plt.xlabel('Carat')
plt.ylabel('Quantity')
plt.show()
```

```python
sns.scatterplot("carat","price", data=diamonds, hue='color2')
plt.title('Price vs Carat')
plt.xlabel('Carat')
plt.ylabel('Price')
plt.show()
```

```python
g = sns.jointplot(
    data=diamonds,
    x="cut2", y="price", 
    kind="kde",
)
```

```python
for feature in diamonds.dtypes[diamonds.dtypes != object].index:
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    sns.violinplot(x= feature, data= diamonds)
    plt.subplot(1, 2, 2)
    sns.boxplot(x= feature, data= diamonds)
    plt.show()
```

```python
sns.set_theme(style="darkgrid")

ax = sns.jointplot(
    data=diamonds,
    x="price", y="carat", hue="cut",
    kind="kde",
    palette='Blues'
)
plt.title("Price vs Carat ")
plt.xlabel("Price")
plt.ylabel("Carat")
```
