```python
import json
import requests
import pandas as pd
```

```python
from stackapi import StackAPI, StackAPIError 

SITE = StackAPI('stackoverflow')

questions = SITE.fetch('questions', fromdate= 1635724800, todate=1638230400, min=1, sort='votes')
#questions = SITE.fetch('questions', fromdate= 1635724800, todate=1638230400, min=20, tagged='python', sort='votes')
questions
```

```python
#print(questions)
items =questions['items']
for it in items:
    print(it['title'])
```

```python
questions_df=pd.DataFrame(questions)
```

```python
import pandas as pd
data = pd.DataFrame(questions)
data.head()
```

```python
def flatten(data, col_list):
    for column in col_list:
        flattened = pd.DataFrame(dict(data[column])).transpose()
        columns = [str(col) for col in flattened.columns]
        flattened.columns = [column + '_' + colname for colname in columns]
        data = pd.concat([data, flattened], axis=1)
        data = data.drop(column, axis=1)
    return data
```

```python
nested_columns = ['items']
```

```python
flat = flatten(questions_df, nested_columns)
```
